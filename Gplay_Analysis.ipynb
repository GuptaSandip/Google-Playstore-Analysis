```python
# dropping unwanted column
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
# impute null values with mean
# m_rating= df['Rating'].mean()

df['Rating'].fillna(df['Rating'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# duplicated values
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Installs'].unique()
```

```python
for i in df.keys():
  print(f'{i} column unique values: \n{df[i].unique()}') # df[i]
  print()
  print('-'*85)
```

```python
def calsize(x):
  if 'M' in x:   # if 'M' in '12M'
    x=x.replace('M','') # '12M' --> '12'
    return float(x) # '12' --> 12.0
  elif 'k' in x:
    x=x.replace('k','')
    return float(x)/1024
  else:
    return np.nan

df['Size'] = df['Size'].map(calsize) # df['Size]
```

```python
df.head()
```

```python
df.info()
```

```python
m_size = df['Size'].mean()
df['Size'].fillna(m_size,inplace=True)
```

```python
df.info()
```

```python
# handling Installs columns
def remove_special_char(x):
  if ',' in x or '+' in x:
    x=x.replace(',','')
    x=x.replace('+','')
    return float(x)
  else:
    np.nan  # by default

df['Installs'] = df['Installs'].map(remove_special_char)
```

```python
df.head()
```

```python
df.info()
```

```python
# handling Price column

def remove_dollar(x):
  if x == '0':
    return float(x)
  elif '$' in x:
    x=x.replace('$','')
    return float(x)
  else:
    np.nan

df['Price'] = df['Price'].map(remove_dollar)
```

```python
df.head()
```

```python
df.info()
```

```python
# handling reviews column
df['Reviews'] = df['Reviews'].astype(float)
```

```python
df.info()
```

```python
for i in df.keys():
  print(f'{i} column unique values: \n{df[i].unique()}') # df[i]
  print()
  print('-'*85)
```

```python
# which category app installs most

data = df.groupby('Category')['Installs'].sum().sort_values(ascending=False)[:5]

import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x= data.index,y=data.values)
plt.xticks(rotation=90)
plt.show()
# print(f'Category:{[i for i in data.index]} , Install:{[i for i in data.values]}')
print(data)
```
